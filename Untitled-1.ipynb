**GRÁFICO 1: HISTOGRAMA**

1. O que é um *histograma?* (Resumo simples)

Mostra como os dados estão distribuídos.

Divide os valores em “faixas” (bins).

Você consegue ver se os dados estão:

**- concentrados**

**- espalhados**

**- simétricos**

**- com caudas**

com vários picos (multimodal)

É o melhor gráfico para entender a distribuição de uma variável numérica.

```python
plt.hist(estudantes['nota_matematica'])
plt.xlabel('notas de matemática')
plt.ylabel('Frequência')
plt.title('Histograma das notas de matemática')
plt.show()
```

**Conclusão do gráfico: Histograma das notas de matemática**

O histograma indica que:

- A maioria dos alunos teve desempenho bom ou muito bom em matemática.

- A quantidade de notas baixas é pequena.

- A matéria parece ter resultados melhores comparada a outras distribuições que normalmente vemos (mais simétricas ou centradas em 60–70).

- As barras sobem conforme a nota aumenta.

Isso significa que mais alunos tiraram notas altas do que notas baixas.

**GRÁFICO 2 — BOXPLOT**

 2. O que é um boxplot? (Resumo)

Ele mostra como os dados estão distribuídos e destaca principalmente:

**- Mediana**

**- Quartis**

**- Dispersão**

**- Outliers** (valores muito fora do padrão)

É excelente para ver variação e extremos.
Ele complementa o histograma:

Histograma → mostra a forma da distribuição

Boxplot → mostra os valores principais e extremos

```python
plt.boxplot(estudantes['nota_matematica'])
plt.title("boxplot das notas de matemática")
plt.ylabel('nota')
plt.show()
```

Conclusão do **boxplot** das notas de matemática

- A mediana está por volta de 78, mostrando que metade dos alunos tirou notas acima desse valor.

- A caixa está mais próxima da parte superior do gráfico, indicando concentração de notas mais altas.

- Existem poucos outliers para baixo, que são alunos com desempenho abaixo do grupo.

- A distribuição é levemente assimétrica, mas sem grandes desvios.

- No geral, o desempenho em matemática é alto e consistente entre os alunos.

**GRÁFICO 3: GRÁFICO DE BARRAS**

O gráfico de barras é um tipo de visualização usado para comparar valores entre categorias.
Cada barra representa uma categoria (como sexo, grupo étnico, turma etc.) e o tamanho da barra mostra a quantidade, média ou proporção daquela categoria.

**-Comparar grupos de forma clara e rápida**

**-Identificar diferenças importantes entre categorias**

**-Visualizar tendências de desempenho ou distribuição**

**-Destacar qual grupo tem maior ou menor valor em uma métrica**

```python
media = estudantes.groupby('sexo')['nota_leitura'].mean()

plt.bar(media.index, media.values)
plt.xlabel('sexo')
plt.ylabel('média das notas de leitura')
plt.title('média das notas de leitura por sexo')
plt.show()
```

Conclusão do **gráfico de barras** — média das notas por sexo

- A média das notas de leitura é praticamente igual entre estudantes do sexo feminino e masculino.

- Isso indica que o sexo não influencia significativamente o desempenho em leitura dentro desse conjunto de dados.

- O mesmo comportamento se repete para as outras matérias (matemática e escrita), mostrando que o desempenho acadêmico é equilibrado entre os sexos.

- Portanto, não há diferença relevante entre os grupos nesse aspecto.

**Gráfico 4 — Scatter Plot** (Dispersão)

4. O que este gráfico mostra

O scatter plot é usado para ver a relação entre duas variáveis numéricas.
Ele ajuda a identificar:

**-tendência** (correlação positiva, negativa ou inexistente)

**-padrões**

**-agrupamentos**

**-pontos fora da curva**

É um dos gráficos mais usados em análise exploratória de dados (EDA).

```python
plt.figure(figsize=(8,5))
plt.scatter(estudantes['nota_matematica'], estudantes['nota_leitura'])
plt.title("Relação entre matemática e leitura")
plt.xlabel('nota de matemática')
plt.ylabel('nota de leitura')
plt.show()
```

Conclusão — **Gráfico de Dispersão** (matemática x leitura)

- Os pontos estão espalhados, mas existe uma leve tendência positiva: conforme a nota de matemática aumenta, a nota de leitura também tende a aumentar.

- Ainda assim, a relação é fraca, pois há muitos pontos dispersos em várias direções.

- Isso mostra que um bom desempenho em matemática não garante bom desempenho em leitura — mas estudantes com notas muito baixas em matemática  normalmente também não têm notas tão altas em leitura.

- O gráfico sugere que as habilidades podem estar moderadamente associadas, já que ambas dependem de compreensão lógica e interpretação.

**Gráfico 5 — Heatmap de Correlação**

5. O que este gráfico faz (resumo simples)

O heatmap mostra o quanto duas ou mais variáveis se relacionam.
Ele usa cores para indicar:

**-Correlação positiva** (próximo de 1) → quando uma variável aumenta, a outra também tende a aumentar.

**-Correlação negativa** (próximo de -1) → quando uma aumenta e a outra diminui.

**-Correlação perto de 0** → pouca ou *nenhuma relação*.

É muito usado em ciência de dados para entender padrões entre variáveis numéricas.

```python
import seaborn as sns
plt.figure(figsize=(8,5))
sns.heatmap(estudantes[['nota_matematica', 'nota_leitura', 'nota_escrita']].corr(),
            annot=True,
            cmap='Blues'
            )

plt.title('Mapa de correlação entre matemática, leitura e escrita')
plt.show()
```

Conclusão — **Mapa de Correlação**

- As notas de leitura e escrita possuem correlação forte (0.83), indicando que alunos que leem bem também escrevem bem — o que faz sentido, pois ambas dependem de interpretação e domínio da língua.

- A relação entre matemática e as outras duas matérias (leitura e escrita) é fraca (0.21 a 0.24), sugerindo que desempenho em matemática não depende diretamente de habilidades de leitura/escrita.

- No geral, o mapa mostra que as áreas linguísticas são fortemente relacionadas entre si, enquanto a matemática se comporta de maneira mais independente.

**Gráfico 6 — Gráfico Pizza**

 6. O que é e para que serve

Um gráfico de pizza (pie chart) é um tipo de visualização que representa um conjunto de dados dividido em partes proporcionais.

Cada fatia da pizza representa uma categoria e seu tamanho mostra quanto aquela categoria contribui para o total geral.

É ideal para responder perguntas como:

**-Qual categoria tem maior participação?**

**-Como os grupos estão distribuídos em porcentagem?**

Esse tipo de gráfico funciona muito bem quando:

**-Existem *poucas categorias* (2 a 6, no máximo).**

**-O objetivo é *comparar proporções*, não valores absolutos.**

A área das fatias ajuda a visualizar rapidamente qual grupo é dominante ou se as categorias estão equilibradas dentro do conjunto. 

```python
contagem = estudantes['grupo_etnico'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(contagem, labels=contagem.index, autopct='%1.1f%%')
plt.title('distribuição de alunos por etnia')
plt.show()
```

Conclusão do gráfico de distribuição por etnia

O gráfico de pizza mostra a distribuição dos alunos entre os diferentes grupos étnicos

- Grupo A possui a maior participação, com cerca de 21% dos alunos.

- Grupos B, C e E aparecem logo em seguida, todos muito próximos, variando entre 20,2% e 20,9%, indicando uma distribuição bastante uniforme.

- Grupo D é o menor grupo, com aproximadamente 17%, mas ainda assim representa uma parcela significativa.

A distribuição torna visível que não há uma concentração excessiva em nenhum grupo específico, o que ajuda a evitar distorções em análises posteriores sobre desempenho, correlação ou métricas comparativas. Essa uniformidade reforça a ideia de um conjunto de dados bem balanceado, útil para estudos *educacionais* e *estatísticos*.
